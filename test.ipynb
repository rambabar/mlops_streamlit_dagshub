```python
import sqlite3
import numpy as np

def get_user_data():
    conn = sqlite3.connect('user_inputs.db')
    cursor = conn.cursor()
    cursor.execute('SELECT * FROM user_data')
    data = cursor.fetchall()
    conn.close()
    return data

def main():
    # Fetching data from SQLite database
    user_data = get_user_data()
    if user_data:
        # Convert to numpy array for prediction
        user_data_np = np.array([list(data[1:]) for data in user_data])
        print("User data for prediction:")
        print(user_data_np)
    else:
        print("No user data available for prediction.")

if __name__ == "__main__":
    main()
```
